```python
df["Commodity"].value_counts().head(10).plot(kind="bar", figsize=(10,5))

plt.title("Top 10 Vegetables")
plt.show()
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["Modal_Price"], bins=30)
plt.title("Modal Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
avg = df.groupby("Commodity")["Modal_Price"].mean().sort_values(ascending=False).head(10)

avg.plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Average Vegetable Prices")
plt.xlabel("Commodity")
plt.ylabel("Average Price")
plt.show()
```
